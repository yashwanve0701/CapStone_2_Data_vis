 Distribution is skewed to the left, it implies more instances with shorter lead times.

#### Chart - 14 - Correlation Heatmap

Positive correlations are represented by warmer colors, while negative correlations are represented by cooler colors. This visualization helps identify which variables are strongly correlated, moderately correlated, or not correlated at all.
